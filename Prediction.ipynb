```python
median_imcome=df['Income'].median()
```

```python
df['Income'] = df['Income'].apply(lambda x: median_imcome if x > 114000 else x)
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median())  # Replaces NaNs with 0
```

```python
df['Income'].iplot(kind="box",boxpoints="outliers")
```

```python
def income_cat(x):
    if x>0 and x<30000:
        return '0-30000'
    elif x>30000 and x<60000 :
        return '30000-60000'
    elif x>60000 and x<90000:
        return '60000-90000'
    else:
        return 'More than 90000'
```

```python
df['income_range']=df['Income'].apply(income_cat)
```

```python
income_range=df['income_range'].value_counts()
```
